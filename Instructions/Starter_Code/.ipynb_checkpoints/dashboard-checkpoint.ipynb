# San Francisco Rental Prices Dashboard

In this notebook, you will compile the visualizations from the previous analysis into functions that can be used for a Panel dashboard.

```python
# imports

import plotly.express as px
import pandas as pd
import hvplot.pandas
import matplotlib.pyplot as plt
import os
import panel as pn
from panel.interact import interact
pn.extension('plotly')
from pathlib import Path
from dotenv import load_dotenv
```

```python
# Read the Mapbox API key
load_dotenv()
mapbox_token = os.getenv("MAPBOX_API_KEY")
px.set_mapbox_access_token(mapbox_token)
```

# Import Data

```python
# Import the CSVs to Pandas DataFrames
file_path = Path("Data/sfo_neighborhoods_census_data.csv")
sfo_data = pd.read_csv(file_path, index_col="year")


file_path = Path("Data/neighborhoods_coordinates.csv")
df_neighborhood_locations = pd.read_csv(file_path)
```

- - -

## Panel Visualizations

In this section, you will copy the code for each plot type from your analysis notebook and place it into separate functions that Panel can use to create panes for the dashboard. 

These functions will convert the plot object to a Panel pane.

Be sure to include any DataFrame transformation/manipulation code required along with the plotting code.

Return a Panel pane object from each function that can be used to build the dashboard.

Note: Remove any `.show()` lines from the code. We want to return the plots instead of showing them. The Panel dashboard will then display the plots.

```python
# Define Panel Visualization Functions
def housing_units_per_year():
    """Housing Units Per Year."""
    
    
    avg_housing_units = sfo_data["housing_units"].groupby('year').mean()
    min_v = avg_housing_units.min()
    max_v = avg_housing_units.max()
    std_v = avg_housing_units.std()
    
    housing_figure =plt.figure()
    
    avg_plot = avg_housing_units.plot.bar(ylim=(min_v-std_v, max_v+std_v))
    avg_plot.set_title("Housing Units in San Francisco from 2010 to 2016", fontsize=14, fontweight="bold")
    avg_plot.set_xlabel("YEAR")
    avg_plot.set_ylabel("Housing Units")

    plt.close(housing_figure)
    
    return pn.pane.Matplotlib(housing_figure)


def average_gross_rent():
    """Average Gross Rent in San Francisco Per Year."""
    avg_gross_rent = sfo_data[["sale_price_sqr_foot","gross_rent"]].groupby('year').mean()
    
    avg_figure = plt.figure()
    
    avg_plot_line = avg_gross_rent["gross_rent"].plot()
    avg_plot_line.set_title("Average Gross Rent in San Francisco", fontsize=14, fontweight="bold")
    avg_plot_line.set_xlabel("Year")
    avg_plot_line.set_ylabel("Housing Units")
    
    plt.close(avg_figure)
    
    return pn.pane.Matplotlib(avg_figure)

def average_sales_price():
    """Average Sales Price Per Year."""
    avg_gross_rent = sfo_data[["sale_price_sqr_foot","gross_rent"]].groupby('year').mean()
    
    avg_sales_figure = plt.figure()
    
    avg_sales_line = avg_gross_rent["sale_price_sqr_foot"].plot()
    avg_sales_line.set_title("Average Sales Price per Year", fontsize=16, fontweight="bold")
    avg_sales_line.set_xlabel("Year")
    avg_sales_line.set_ylabel("Avg Sale Price")
    
    plt.close(avg_sales_figure)
    
    return pn.pane.Matplotlib(avg_sales_figure)


def average_price_by_neighborhood():
    """Average Prices by Neighborhood."""
    avg_price_hoods = sfo_data.groupby(['year','neighborhood']).mean()
    avg_price_hoods.reset_index(inplace=True)
    avg_price_hoods.head(10)
    plot_hood = avg_price_hoods.hvplot.line(x="year", y="sale_price_sqr_foot", groupby="neighborhood")

    return plot_hood


def top_most_expensive_neighborhoods():
    """Top 10 Most Expensive Neighborhoods."""
    expensive_hoods=sfo_data.groupby('neighborhood').mean()
    
    expensive_hoods.sort_values(by = 'sale_price_sqr_foot',ascending=False, inplace=True)
    expensive_hoods = expensive_hoods.reset_index()
    expensive_hoods2 = expensive_hoods.head(10)
    expensive_hoods_plot = expensive_hoods2.hvplot.line(x="neighborhood", y="sale_price_sqr_foot",rot=90)

    return   expensive_hoods_plot

def parallel_coordinates():
    """Parallel Coordinates Plot."""
    expensive_hoods=sfo_data.groupby('neighborhood').mean()
    expensive_hoods.sort_values(by = 'sale_price_sqr_foot',ascending=False, inplace=True)
    expensive_hoods = expensive_hoods.reset_index()
    expensive_hoods2 = expensive_hoods.head(10)
    parr_corr = px.parallel_coordinates(expensive_hoods2, color='sale_price_sqr_foot',color_continuous_scale = px.colors.sequential.Inferno)

    return parr_corr

def parallel_categories():
    """Parallel Categories Plot."""
    expensive_hoods=sfo_data.groupby('neighborhood').mean()
    expensive_hoods.sort_values(by = 'sale_price_sqr_foot',ascending=False, inplace=True)
    expensive_hoods = expensive_hoods.reset_index()
    expensive_hoods2 = expensive_hoods.head(10)
    parr_cat = px.parallel_categories(
    expensive_hoods2,
    dimensions=["neighborhood","sale_price_sqr_foot","housing_units","gross_rent"],
    color="sale_price_sqr_foot",color_continuous_scale=px.colors.sequential.Inferno,
    )
    return parr_cat

def neighborhood_map():
    """Neighborhood Map"""
    
    df_neighborhood_locations.rename(columns = {'Neighborhood' : 'neighborhood'}, inplace=True)
    mean_values = sfo_data.groupby('neighborhood').mean()
    mean_values.reset_index(inplace=True)
    dropped_neighborhood = pd.merge(df_neighborhood_locations, mean_values, on='neighborhood')
    fig = px.scatter_mapbox(dropped_neighborhood, lat="Lat", lon="Lon", color="gross_rent", size="sale_price_sqr_foot",title="Averange Sale Price Per Square Foot and Gross Rent in San Francisco",
    color_continuous_scale=px.colors.cyclical.IceFire, size_max=15, zoom=4)
    return fig
```

## Panel Dashboard

In this section, you will combine all of the plots into a single dashboard view using Panel. Be creative with your dashboard design!

```python
# YOUR CODE HERE!
Welcome = pn.pane.Markdown(
"""
##Welcome!

The data in the following tabs provide visualtions powered by PyViz to illustrate select property information of the San Francisco Area between 2010 and 2016.
You can select from any of the following tabs to explore different types of analysis.

1. Welcome - Provides an intro of the data being analyzed.
2. Housing Rental Analysis 2010 - 2016, containing the following charts:
    - Housing Units Per Year
    - Average Gross Rent Per Year
    - Average Sales Price Per Year
3. Neighborhood Analysis 2010 - 2016, containing the following charts:
    - Average Prices by Neighborhood  
    - Top 10 Most Expensive Neighborhoods  
4. Parallel Coordinates and Parallel Categories Analysis
    - Average values per neighborhood - Parallel Coordinates  
    - Average values per neighborhood - Parallel Categories  
5. Geospatial Analysis, containting an interactive map with data on average prices per neighborhood.
"""
)
```

```python
tab_1 = pn.Column(
    ("# San Francisco Housing Rental Analysis - 2010 to 2016"),
    ("## Housing Units Per Year"),housing_units_per_year(),
    ("## Average Gross Rent Per Year"),average_gross_rent(),
    ("## Average Sales Price Per Year"),average_sales_price())
```

```python
tab_2 = pn.Column(
    ("## Average Prices by Neighborhood"),average_price_by_neighborhood,
    ("## Top 10 Most Expensive Neighborhoods"),top_most_expensive_neighborhoods())

tab_3 = pn.Column(
    ("# Parallel Coordinates and Parallel Categories Analysis"),
    ("## Parallel Coordinates Plot"),parallel_coordinates,
    ("## Parallel Categories Plot"),parallel_categories())

tab_4 = pn.Column(
     ("# Averange Sale Price Per Square Foot and Gross Rent in San Francisco"),neighborhood_map())
```

## Serve the Panel Dashboard

```python
sf_panel = pn.Tabs(
    ("Welcome",Welcome),
    ("Rental Analysis", tab_1),
    ("Neighborhood Analysis", tab_2),
    ("Parallel Analysis",tab_3),
    ("Geospatial Analysis", tab_4)
)

```
